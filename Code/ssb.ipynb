```python
import pandas as pd
```

```python
#load the data
divorcedata = pd.read_csv('skillsmisser_agegroups.csv')

```

```python
print(divorcedata.head())

# Combine age groups into broader intervals
divorcedata['AgeGroup'] = divorcedata['AgeGroup'].replace({
    '15-19': '15-19',
    '20-24': '20-29',
    '25-29': '20-29',
    '30-34': '30-39',
    '35-39': '30-39',
    '40-44': '40-49',
    '45-49': '40-49',
    '50-54': '50-59',
    '55-59': '50-59',
    '60-64': '60-69',
    '65-69': '60-69',
    '70+': '70+'
})

# Group by age group and average divorce rates
divorcedata = divorcedata.groupby('AgeGroup')['Skillsmisser'].mean().reset_index()

print(divorcedata.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


# Create figure
plt.figure(figsize=(10, 6))

# Colorblind-safe blue
ax = sns.barplot(
    x='AgeGroup',
    y='Skillsmisser',
    data=divorcedata,
    color='#0072B2'
)

# Add values above bars
for container in ax.containers:
    ax.bar_label(
        container,
        fmt='%.1f',
        fontsize=9,
        padding=3
    )

# Labels and title
plt.xlabel('Age Group', fontsize=12)
plt.ylabel('Approximate Combined Divorce Rate Per 1000', fontsize=12)

plt.title(
    'Approximate combined divorce rates across Age Groups per 1000 individuals separated in Norway (2025)' ,
    fontsize=14
)

plt.tight_layout()
plt.show()
```
